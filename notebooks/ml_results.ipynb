#### Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import sys
from datetime import datetime
from scipy import stats
```

```python
def WriteJSON(obj,filename):
    with open(filename, 'w+') as outfile:
        try:
            obj_json = json.dumps(obj, sort_keys=True, indent=4,default=str)
            outfile.write(obj_json)
        except Exception as e:
            print(e, file=sys.stderr)
            print('File not written.')
```

```python
def ReadJSON(filename):
    obj = []
    try: 
        with open(filename, 'r') as infile:
            obj = json.load(infile)
    except Exception as e:
        print(e, file=sys.stderr)
        print('File not found.')
        
    return obj
```

### Start Program

```python
filename = '../data/results/results.json'
results = pd.read_json(filename)

results = results.drop(index=[13])

counts = results['Counts']
Counts_0 = []
Counts_1 = []
for count in counts:
    Counts_0.append(count['0'])
    Counts_1.append(count['1'])

results['Counts_0'] = Counts_0
results['Counts_1'] = Counts_1

results = results.drop(columns=['Counts'])

allclfs = results['classifiers']
BGN = []
DTC = []
KNN = []
SVM = []
collector = [BGN,DTC,KNN,SVM]
i = 0
for clasifiers in allclfs:
    for clf in clasifiers:
        tmp = list(clf.keys())[0]
        collector[i].append(clf[tmp]['Report']['accuracy'])
        if i < 3:
            i += 1
        else:
            i = 0
        
results['BGN-Accu'] = collector[0]
results['DTC-Accu'] = collector[1]
results['KNN-Accu'] = collector[2]
results['SVM-Accu'] = collector[3]


allclfs = results['classifiers']

BGNTN = []
BGNFP = []
BGNFN = []
BGNTP = []

DTCTN = []
DTCFP = []
DTCFN = []
DTCTP = []

KNNTN = []
KNNFP = []
KNNFN = []
KNNTP = []

SVMTN = []
SVMFP = []
SVMFN = []
SVMTP = []
for clasifiers in allclfs:
    for clf in clasifiers:
        tmp = list(clf.keys())[0]
        tmp2 = clf[tmp]['Matrix']
#         print(tmp)
        if tmp == 'BGN':
            tn = int(list(tmp2)[0][1:3])
            BGNTN.append(tn)
            fp = int(list(tmp2)[0][4:6])
            BGNFP.append(fp)
            fn = int(list(tmp2)[1][1:3])
            BGNFN.append(fn)
            tp = int(list(tmp2)[1][4:6])
            BGNTP.append(tp)
        elif tmp == 'DTC':
            tn = int(list(tmp2)[0][1:3])
            DTCTN.append(tn)
            fp = int(list(tmp2)[0][4:6])
            DTCFP.append(fp)
            fn = int(list(tmp2)[1][1:3])
            DTCFN.append(fn)
            tp = int(list(tmp2)[1][4:6])
            DTCTP.append(tp)
        elif tmp == 'KNN':
            tn = int(list(tmp2)[0][1:3])
            KNNTN.append(tn)
            fp = int(list(tmp2)[0][4:6])
            KNNFP.append(fp)
            fn = int(list(tmp2)[1][1:3])
            KNNFN.append(fn)
            tp = int(list(tmp2)[1][4:6])
            KNNTP.append(tp)
        elif tmp == 'SVM':
            tn = int(list(tmp2)[0][1:3])
            SVMTN.append(tn)
            fp = int(list(tmp2)[0][4:6])
            SVMFP.append(fp)
            fn = int(list(tmp2)[1][1:3])
            SVMFN.append(fn)
            tp = int(list(tmp2)[1][4:6])
            SVMTP.append(tp)


results['BGN-TN'] = BGNTN
results['BGN-FP'] = BGNFP
results['BGN-FN'] = BGNFN
results['BGN-TP'] = BGNTP

results['DTC-TN'] = DTCTN
results['DTC-FP'] = DTCFP
results['DTC-FN'] = DTCFN
results['DTC-TP'] = DTCTP

results['KNN-TN'] = KNNTN
results['KNN-FP'] = KNNFP
results['KNN-FN'] = KNNFN
results['KNN-TP'] = KNNTP

results['SVM-TN'] = SVMTN
results['SVM-FP'] = SVMFP
results['SVM-FN'] = SVMFN
results['SVM-TP'] = SVMTP

results = results.drop(columns=['classifiers'])

results
```
